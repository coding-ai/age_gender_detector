```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('/content/gdrive/MyDrive/Keras/Kodlar ve görüntüler/Gürültü giderme/8 - Gürültü giderme/median.jpg')
```

```python
img1 = cv.medianBlur(img,9)
```

```python
basliklar = ['Orijinal', 'Median Blur']
images = [img, img1]
```

```python
plt.figure(figsize=(15,20))
for i in range(2):
  plt.subplot(1,2,i+1), plt.imshow(images[i]), plt.title(basliklar[i])
  plt.xticks([]), plt.yticks([])

plt.show()
```
