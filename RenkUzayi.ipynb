```python
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('/content/gdrive/MyDrive/Keras/Kodlar ve görüntüler/Renk uzayì uygulama/5 - 2 - Renk uzayì uygulama/cube.png')
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(15,20))

# OpenCv BGR
plt.subplot(1,2,1)
plt.title('OpenCv BGR')
plt.imshow(img)
plt.xticks([]), plt.yticks([])

# Matplotlib RGB
plt.subplot(1,2,2)
plt.title('Matplotlib RGB')
plt.imshow(rgb_img)
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
red = rgb_img.copy()
red[:,:, 1]=0
red[:,:,2]=0

green=rgb_img.copy()
green[:,:,0]=0
green[:,:,2]=0

blue=rgb_img.copy()
blue[:,:,0]=0
blue[:,:,1]=0
```

```python
plt.figure(figsize=(15,20))

plt.subplot(1,3,1)
plt.title('RED')
plt.imshow(red)
plt.xticks([]), plt.yticks([])

plt.subplot(1,3,2)
plt.title('GREEN')
plt.imshow(green)
plt.xticks([]),plt.yticks([])

plt.subplot(1,3,3)
plt.title('BLUE')
plt.imshow(blue)
plt.xticks([]), plt.yticks([])

plt.show()
```
