```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
resim = cv2.imread('/content/gdrive/MyDrive/Keras/Kodlar ve görüntüler/Görüntüde geometrik iƒlemler/4 - 4 - Görüntüde geometrik iƒlemler/messi.jpg')
```

```python
plt.imshow(resim), plt.title('Orijinal')
```

```python
genislik,uzunluk = resim.shape[:2]
```

```python
M = np.float32([[1,0,50],[0,1,50]])
kaydirma =cv2.warpAffine(resim,M,(uzunluk,genislik))
```

```python
plt.subplot(122), plt.imshow(kaydirma), plt.title('kaydırılmış')
plt.show()
```

```python
satir, sutun = resim.shape[:2]
```

```python
M = cv2.getRotationMatrix2D((sutun/2,satir/2), 45, 1)
cevirme = cv2.warpAffine(resim, M,(sutun,satir))
```

```python
plt.subplot(121), plt.imshow(resim), plt.title('Orijinal')
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(cevirme), plt.title('Çevrilmiş')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
img = cv2.imread('/content/gdrive/MyDrive/Keras/Kodlar ve görüntüler/Görüntüde geometrik iƒlemler/4 - 4 - Görüntüde geometrik iƒlemler/affine.jpg')
```

```python
satir, sutun = img.shape[:2]
```

```python
point1 = np.float32([[50,50],[200,50],[50,200]])
point2 = np.float32([[10,100],[200,50],[100,250]])
```

```python
M = cv2.getAffineTransform(point1,point2)
```

```python
hassas = cv2.warpAffine(img, M, (sutun, satir))
```

```python
plt.subplot(121), plt.imshow(img), plt.title('Orijinal2')
plt.subplot(122), plt.imshow(hassas), plt.title('Hassas Dönüşüm')
plt.show()
```

```python
foto = cv2.imread('/content/gdrive/MyDrive/Keras/Kodlar ve görüntüler/Görüntüde geometrik iƒlemler/4 - 4 - Görüntüde geometrik iƒlemler/perspective.jpg')
```

```python
plt.subplot(121), plt.imshow(foto), plt.title('Orijinal3')
plt.show()
```

```python
pts1 = np.float32([[25,25],[185,25],[10,195],[195,195]])
pts2 = np.float32([[0,0],[200,0],[0,200],[200,200]])
```

```python
M = cv2.getPerspectiveTransform(pts1,pts2)
```

```python
pers = cv2.warpPerspective(foto, M, (200,200))
```

```python
plt.subplot(121), plt.imshow(foto), plt.title('Orijinal3')
plt.subplot(122), plt.imshow(pers), plt.title('Perspektif Dönüşüm')
plt.show()
```
