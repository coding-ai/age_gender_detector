```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('/content/gdrive/MyDrive/Keras/Kodlar ve görüntüler/Görüntü yumuƒatma/7 - Görüntü yumuƒatma/filter.jpg')
```

```python
kernel = np.ones((5,5), np.float)/25
img2 = cv.filter2D(img,-1,kernel)
```

```python
plt.subplot(121), plt.imshow(img), plt.title('Orijinal')
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(img2), plt.title('Ortalama Filtre')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
img3 = cv.GaussianBlur(img,(5,5),0)
```

```python
basliklar = ['Orijina', 'Ortalama filte', 'Gauss Blur Filtre']
fotolar = [img, img2, img3]
```

```python
plt.figure(figsize=(15,20))
for i in range(3):
  plt.subplot(1,3,i+1), plt.imshow(fotolar[i])
  plt.title(basliklar[i])
  plt.xticks([]), plt.yticks([])

plt.show()
```
