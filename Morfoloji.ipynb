```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('/content/gdrive/MyDrive/Keras/Kodlar ve görüntüler/Morfolojik iƒlemler/6 - Morfolojik iƒlemler/j.png', 0)
kernel = np.ones((5,5),np.uint8)
```

```python
asindirma = cv.erode(img, kernel, iterations=1)
genisletme = cv.dilate(img, kernel,iterations=1)
```

```python
basliklar = ['Orijinal', 'Aşındırma', 'Genişletme']
resimler = [img, asindirma, genisletme]
```

```python
plt.figure(figsize=(10,12))
for i in range(3):
  plt.subplot(1,3,i+1), plt.imshow(resimler[i])
  plt.title(basliklar[i])
  plt.xticks([]), plt.yticks([])

plt.show()
```
